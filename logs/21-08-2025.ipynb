### Problem 1:  Wave Array
Given an array of integers A, sort the array into a wave-like array and return it.
In other words, arrange the elements into a sequence such that

a1 >= a2 <= a3 >= a4 <= a5..... 
NOTE: If multiple answers are possible, return the lexicographically smallest one.

```python
# Paste your solution here
class Solution:
    # @param A : integer
    def solve(self, A):
        A.sort()
        for i in range(0, len(A) - 1, 2):
            A[i], A[i + 1] = A[i + 1], A[i]
        return A


Solution().solve([5, 2, 8])
```

### Problem 2: B Closest Points to Origin
You are developing a feature for Zomato that helps users find the nearest restaurants to their current location. It uses GPS to determine the user's location and has access to a database of restaurants, each with its own set of coordinates in a two-dimensional space representing their geographical location on a map. The goal is to identify the "B" closest restaurants to the user, providing a quick and convenient way to choose where to eat.

Given a list of restaurant locations, denoted by A (each represented by its x and y coordinates on a map), and an integer B representing the number of closest restaurants to the user. The user's current location is assumed to be at the origin (0, 0).

Here, the distance between two points on a plane is the Euclidean distance.

You may return the answer in any order. The answer is guaranteed to be unique (except for the order that it is in.)

NOTE: Euclidean distance between two points P1(x1, y1) and P2(x2, y2) is sqrt( (x1-x2)2 + (y1-y2)2).

```python
# Paste your solution here
class Solution:
    def euclidean_distance(self, x, y):
        return pow(x, 2) + pow(y, 2)

    # @param A : integer
    def solve(self, A, B):
        res = {}
        for i in range(len(A)):
            distance = self.euclidean_distance(A[i][0], A[i][1])
            if distance in res:
                res[distance].append(A[i])
            else:
                res[distance] = [A[i]]
        keys = list(res.keys())
        keys.sort()
        ans = []
        cnt = 0
        i = 0
        while cnt < B and i < len(keys):
            for v in res[keys[i]]:
                ans.append(v)
            cnt += len(res[keys[i]])
            i += 1
        return ans


Solution().solve(
    [
        [50, 35],
        [6, 4],
        [1, 26],
        [35, 30],
        [21, 14],
        [16, 11],
        [50, 35],
        [22, 37],
        [26, 3],
        [96, 74],
        [78, 63],
        [82, 106],
        [91, 107],
        [62, 107],
        [85, 82],
        [74, 69],
        [66, 105],
        [109, 73],
        [76, 108],
        [63, 64],
        [81, 104],
        [91, 106],
        [68, 60],
        [69, 65],
        [86, 67],
    ],
    9,
)
```

### Problem 3: Tens Digit Sorting
Given an array A of N integers. Sort the array in increasing order of the value at the tens place digit of every number.

If a number has no tens digit, we can assume value to be 0.
If 2 numbers have same tens digit, in that case number with max value will come first
Solution should be based on comparator.

```python
from functools import cmp_to_key


# Paste your solution here
class Solution:
    def tens_digit(self, A):
        return (A // 10) % 10

    def comparator(self, A, B):
        if self.tens_digit(A) > self.tens_digit(B):
            return 1
        elif self.tens_digit(A) < self.tens_digit(B):
            return -1
        else:
            return B - A

    # @param A : integer
    def solve(self, A):
        A.sort(key=cmp_to_key(self.comparator))
        return A


Solution().solve([888,417,136,979,360,711,569,649,693,763,586,259,715,892,366,309,170,72,992,483,245,5,421,66,466,368,109,774,716,89,630,154,68,749,857,699,68,539,467,69,531,888,46,364,700,827,181,44,470,145,859,819,245,618,957,737,232,394,435,447,729,209,291,117,824,939,646,242,936,145,369,799,593,583,142,660,156,84,776,866,482,229,722,848,838,809,241,559,342,830,173,910,158,251,829,422,268,148,616,998,949,770,284,760,135,608,722,506,685,359,992,95,927,882,470,60,71,99,354,851,124,100,348,187,467,775,378,237,193,903,685,785,378,261,362,463,389,420,59,95,510,628,99,89,315,611,118,494,632,663,366,389,555,965,183,294,328,262,12,99,182,908,167,98,15,370,693,661,956,649,784,281,911,740,379,962,504,61,154,654,841,245,383,278,194,888,102,552,493,758,442,127,866,204,989,57,637,353,860,896,531,433,209,314,753,86,240,625,365,66,506,732,435,205,653,764,241,813,121,872,627,313,396,385,992,534,344,503,587,298,976,298,715,540,222,593,675,290,40,879,426,560,718,137,573,360,414,537,944,413,801,238,33,831,779,66,781,890,454,258,753,502,871,602,810,952,776,847,197,600,349,946,616,554,823,612,655])
```

### Problem 4: Sort 01
Given an array A of 0s and 1s of length N. Sort this array.

Note - Do not use any sorting algorithm or library's sort method.

```python
# Paste your solution here
class Solution:
    # @param A : integer
    def solve(self, A):
        cnt = 0
        for i in A:
            if i == 0:
                cnt += 1
        for i in range(cnt):
            A[i] = 0
        for i in range(cnt, len(A)):
            A[i] = 1
        return A

Solution().solve([0,0,1,0,1,1,0])
```
