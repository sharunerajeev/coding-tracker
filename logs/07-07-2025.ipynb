```python
def rotate_matrix(A):
    n = len(A)
    # Transpose the matric
    for i in range(n):
        for j in range(i):
            A[i][j], A[j][i] = A[j][i], A[i][j]
    # Reverse the elements in each row
    for i in range(n):
        A[i] = A[i][::-1]
    return A


rotate_matrix([[1, 2, 3], [4, 5, 6], [7, 8, 9]])

```
```python
def length_of_longest_consecutive_ones(A):
    n = len(A)
    left = [0] * n
    right = [0] * n

    # Build left array
    if A[0] == "1":
        left[0] = 1
    else:
        left[0] = 0
    for i in range(1, n):
        if A[i] == "1":
            left[i] = left[i - 1] + 1
        else:
            left[i] = 0

    # Build right array
    if A[n - 1] == "1":
        right[n - 1] = 1
    else:
        right[n - 1] = 0
    for i in range(n - 2, -1, -1):
        if A[i] == "1":
            right[i] = right[i + 1] + 1
        else:
            right[i] = 0

    # Total number of 1s
    total_ones = 0
    for i in A:
        if i == "1":
            total_ones += 1

    lcs = 0  # Longest consecutive ones
    for i in range(n):
        if A[i] == "0":
            # Edge cases
            if i > 0:
                left_count = left[i - 1]
            else:
                left_count = 0
            if i < n - 1:
                right_count = right[i + 1]
            else:
                right_count = 0
            merged = left_count + right_count

            if merged < total_ones:
                merged += 1

            lcs = max(merged, lcs)
        
    # Edge case when there is no 0s
    if lcs == 0:
        lcs = total_ones

    return lcs


print(length_of_longest_consecutive_ones("1101110"))

```
```python
def transpose(A):
    n = len(A)
    m = len(A[0])
    B = [[0] * n for _ in range(m)]
    for i in range(m):
        for j in range(n):
            B[i][j] = A[j][i]
    return B


print(
    transpose(
        [
            [24, 63, 39, 81, 84, 30],
            [21, 64, 95, 6, 88, 73],
            [33, 6, 63, 96, 86, 66],
            [62, 88, 23, 52, 94, 77],
            [81, 58, 74, 18, 16, 25],
            [26, 40, 88, 64, 72, 23],
            [45, 44, 86, 92, 50, 26],
            [64, 34, 83, 26, 29, 68],
            [43, 42, 7, 17, 45, 52],
            [94, 25, 62, 19, 95, 77],
        ]
    )
)

```
```python
def add_matrices(A, B):
    n = len(A)
    m = len(A[0])
    for i in range(n):
        for j in range(m):
            A[i][j] = A[i][j] + B[i][j]
    return A


print(
    add_matrices([[1, 2, 3], [4, 5, 6], [7, 8, 9]], [[9, 8, 7], [6, 5, 4], [3, 2, 1]])
)

```
```python
def check_anagrams(A, B):
    letters_in_A = {}
    letters_in_B = {}

    for i in A:
        if i in letters_in_A:
            letters_in_A[i] += 1
        else:
            letters_in_A[i] = 1

    for i in B:
        if i in letters_in_B:
            letters_in_B[i] += 1
        else:
            letters_in_B[i] = 1

    for i in letters_in_A:
        if i not in letters_in_B:
            return 0
        if letters_in_A[i] != letters_in_B[i]:
            return 0
        
    return 1


print(check_anagrams('cat', 'bat'))

```
```python
def n_3_repeat_number(A):
    n = len(A)
    a_map = {}
    for i in A:
        if i in a_map:
            a_map[i] += 1
        else:
            a_map[i] = 1
    for i in a_map:
        if a_map[i] > n // 3:
            return i

    return -1


print(n_3_repeat_number([1000441, 1000441, 1000994]))

```
```python
def colorful_number(A):
    A = str(A).strip()
    n = len(A)
    substr = []
    for i in range(n):
        for j in range(i, n):
            substr.append(A[i:j + 1])
    substr_products = []
    for i in substr:
        product = 1
        for j in i:
            product *= int(j)
        substr_products.append(product)
    substr_products.sort()
    for i in range(1, len(substr_products)):
        if substr_products[i-1] == substr_products[i]:
            return 0
    return 1

def optimised_colorful_number(A):
    A = str(A).strip()
    n = len(A)
    seen = set(A)

    for i in range(n):
        product = 1
        for j in range(i, n):
            product *= int(A[j])
            if product in seen:
                return 0
            seen.add(product)

    return 1

print(colorful_number(1542))
print(optimised_colorful_number(1542))
```
